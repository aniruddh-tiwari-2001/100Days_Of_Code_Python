### Conditional selection in Pandas Dataset

```python
df.Sex == 'male'
```

```python
# Find all the passengers who are male
df.loc[df.Sex == 'male']
```

```python
# Find all the passengers who are male and below 18
df.loc[(df.Sex == 'male') & (df.Age<18)]
```

```python
# Find all the passengers who are male or 18 in age
df.loc[(df.Sex=='male') | (df.Age == 18)]
```

```python
# Find how many passengers Embarked 'Q'
df.loc[df.Embarked.isin(['Q'])]
```

### FInding all not null values

```python
df.loc[df.Cabin.notnull()]
```

### FInding all null values

```python
df.loc[df.Cabin.isnull()]
```
